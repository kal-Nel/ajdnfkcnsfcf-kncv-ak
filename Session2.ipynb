```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
x= np.arange(1,4.5)
```

```python
x
```

```python
y = 2.5*x
```

```python
y
```

```python
plt.plot(x,y)
plt.show
```

```python
x
```

```python
y1 = 2*x
```

```python
y2 = 3*x
```

```python
plt.plot(x,y1,color='b',linestyle=':')
plt.plot(x,y2,color='r',linewidth=2)
plt.title('Two Shots in 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color='g',linestyle='-',linewidth=1)
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(x,y2,color='b',linestyle=':',linewidth=2)
plt.grid(True)
plt.show
```
